```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
exchange_rates = pd.read_excel('Data/Processed/DAILY_EXCHANGE_RATES.xlsx')

# Ensure the date is in the correct datetime format (if not already)
exchange_rates['Date'] = pd.to_datetime(exchange_rates['Date'])

# Filter the necessary currencies (assuming we are using Mid values as exchange rate values)
necessary_currencies = ['USDNOK', 'EURNOK', 'CHFNOK', 'GBPNOK']
filtered_data = exchange_rates[exchange_rates['Ticker'].isin(necessary_currencies)]

# Calculate the mean exchange rate per day for a broader perspective
mean_rates = filtered_data.groupby('Date')['Mid'].mean()

# Prepare the plot
plt.figure(figsize=(16, 9))

# Plotting the mean rate
plt.plot(mean_rates.index, mean_rates.values, label='Mean Exchange Rate', color='purple', linewidth=2)

plt.title('Mean Development of NOK Exchange Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Mean Exchange Rate')
plt.legend(title='Ticker')
plt.grid(True)
plt.show()
```

### 1.4 Comparing commodities with energy companies

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
stock_prices = pd.read_excel('Data/Processed/DAILY_STOCK_PRICES.xlsx')
commodity_prices = pd.read_excel('Data/Processed/DAILY_COMMODITY_PRICES.xlsx')

# Ensure the date is in the correct datetime format
stock_prices['Date'] = pd.to_datetime(stock_prices['Date'])
commodity_prices['Date'] = pd.to_datetime(commodity_prices['Date'])

# Define the stocks and commodities to plot, with specific colors
stocks = {
    'EQNR-NO': 'purple',
    'SHEL-GB': 'pink',
    'AKERBP-NO': 'violet'
}
commodities = {
    'OILDB-FDS': 'red',
    #'HHGAS-FDS': 'blue',
    #'AECOCAD-FDS': 'green'
}



# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(16, 9))

# Plot each stock on ax1
for stock, color in stocks.items():
    data = stock_prices[stock_prices['Ticker'] == stock]
    ax1.plot(data['Date'], data['Price'], label=f'Stock: {stock}', color=color)

# Set labels and title
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price')
ax1.set_title('Stock and Commodity Prices Over Time')

# Create a second y-axis for the commodities
ax2 = ax1.twinx()
for commodity, color in commodities.items():
    data = commodity_prices[commodity_prices['Ticker'] == commodity]
    ax2.plot(data['Date'], data['Price'], label=f'Commodity: {commodity}', linestyle='-', color=color)

# Set secondary y-axis label
ax2.set_ylabel('Commodity Price')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Add grid and show plot
ax1.grid(True)
plt.show()
```

### 1.6 Comparing oil price, NOK and stock prices

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
stock_prices = pd.read_excel('Data/Processed/DAILY_STOCK_PRICES.xlsx')
commodity_prices = pd.read_excel('Data/Processed/DAILY_COMMODITY_PRICES.xlsx')
exchange_rates = pd.read_excel('Data/Processed/DAILY_EXCHANGE_RATES.xlsx')

# Ensure the date is in the correct datetime format
stock_prices['Date'] = pd.to_datetime(stock_prices['Date'])
commodity_prices['Date'] = pd.to_datetime(commodity_prices['Date'])
exchange_rates['Date'] = pd.to_datetime(exchange_rates['Date'])

# Calculate the mean exchange rate from the necessary currencies
necessary_currencies = ['USDNOK', 'EURNOK', 'CHFNOK', 'GBPNOK']
filtered_rates = exchange_rates[exchange_rates['Ticker'].isin(necessary_currencies)]
mean_rates = filtered_rates.groupby('Date')['Mid'].mean()

# Define the stocks and commodities to plot, with specific colors
stocks = {
    'EQNR-NO': 'purple',
    'SHEL-GB': 'pink',
    'AKERBP-NO': 'violet'
}
commodities = {'OILDB-FDS': 'red'}

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(16, 9))

# Plot each stock on ax1
for stock, color in stocks.items():
    data = stock_prices[stock_prices['Ticker'] == stock]
    ax1.plot(data['Date'], data['Price'], label=f'Stock: {stock}', color=color)

# Set labels and title
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price', color='tab:blue')
ax1.set_title('Stock Prices, Oil Prices, and Mean Exchange Rate Over Time')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis for the commodities
ax2 = ax1.twinx()
data = commodity_prices[commodity_prices['Ticker'] == 'OILDB-FDS']
ax2.plot(data['Date'], data['Price'], label='Commodity: OILDB-FDS', linestyle='-', color='red')
ax2.set_ylabel('Commodity Price', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Plot the mean exchange rate on a second y-axis
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Offset the right spine of ax3
ax3.plot(mean_rates.index, mean_rates, label='Mean NOK Exchange Rate', color='green', linestyle=':')
ax3.set_ylabel('Mean Exchange Rate', color='green')
ax3.tick_params(axis='y', labelcolor='green')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
ax3.legend(loc='center right')

# Show plot
plt.show()
```
